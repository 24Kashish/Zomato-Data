```python
exchange_rates = {"Indian Rupees(Rs.)": 0.012, "Dollar($)": 1.0, "Pounds(£)": 1.27, "Brazilian Real(R$)": 0.20, "Emirati Diram(AED)": 0.27, "Rand(R)": 0.054, "NewZealand($)": 0.62, "Turkish Lira(TL)": 0.037, "Botswana Pula(P)": 0.074, "Indonesian Rupiah(IDR)": 0.000065, "Qatari Rial(QR)": 0.27, "Sri Lankan Rupee(LKR)": 0.0031}
```

```python
Avg_cost_for_2_USD = []

for i in range(len(data)):
    cur= data.loc[i,"Currency"]
    if cur in exchange_rates:
        cost= data.loc[i, "Average Cost for two"]* exchange_rates[cur]
    Avg_cost_for_2_USD. append(cost)
```

```python
data['Average Cost for two USD'] = Avg_cost_for_2_USD
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe= OrdinalEncoder()
data["Restaurant Name"]= oe.fit_transform(data[["Restaurant Name"]])
data["Cuisines"]= oe.fit_transform(data[["Cuisines"]])
data["Has Table booking"]= oe.fit_transform(data[["Has Table booking"]])
data["Has Online delivery"]= oe.fit_transform(data[["Has Online delivery"]])
data["Is delivering now"]= oe.fit_transform(data[["Is delivering now"]])
data["Country"]= oe.fit_transform(data[["Country"]])
data["Rating text"]= oe.fit_transform(data[["Rating text"]])
```

```python
y= data["Rating text"].to_numpy()
data.drop(columns= "Rating text", inplace= True)
```

```python
x= data.to_numpy()
```

```python
from sklearn. model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size= 0.2, random_state= 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score

lr= LogisticRegression()
svc= SVC()
gnb= GaussianNB()
rvc= RandomForestClassifier()
dc= DecisionTreeClassifier()
abc= AdaBoostClassifier()

mod= [lr, svc, gnb, rvc, dc, abc]

for i in mod:
    i.fit(x_train, y_train)
    y_pred= i.predict(x_test)
    acc= accuracy_score(y_test, y_pred)
    print(f"{i} Accuracy: {acc*100}%")
```
